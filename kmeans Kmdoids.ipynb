```python
# KMeans clustering
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv('Iris.csv')
iris.columns
```

```python
iris = iris.drop(columns=['Id', 'Species'])
```

```python
iris_scaled = StandardScaler().fit_transform(iris)
```

```python
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit_predict(iris)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.show()
```

```python
kmeans_3 = KMeans(n_clusters=3, random_state=42)
iris['Cluster_3'] = kmeans_3.fit_predict(iris_scaled)

sns.scatterplot(x=iris['SepalLengthCm'], y=iris['PetalLengthCm'], hue=iris['Cluster_3'], palette='Set1')
plt.plot()
```

```python
#KMedoid
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn_extra.cluster import KMedoids
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
cc_data = pd.read_csv('Credit_Card.csv')
cc_data.columns


cc_data = cc_data.drop(columns=['CUST_ID'])

cc_data.isna().sum()

cc_data = cc_data.fillna(cc_data.mean())

cc_data.isna().sum()
```

```python
s_score = []
K = range(2, 11)

for k in K:
  kmedoid = KMedoids(n_clusters=k, random_state=42)
  cluster_labels = kmedoid.fit_predict(cc_data)
  s_scores = silhouette_score(cc_data, cluster_labels)

  s_score.append(s_scores)

plt.plot(K, s_score, 'bo-')
plt.show()
```

```python
kmedoid_2 = KMedoids(n_clusters=2, random_state=42)
cc_data['Cluster_2'] = kmedoid_2.fit_predict(cc_data)

sns.scatterplot(x=cc_data['BALANCE'] , y=cc_data['PURCHASES'], hue = cc_data['Cluster_2'])
plt.show()
```
